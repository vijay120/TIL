```python
# Standard python numerical analysis imports:
import numpy as np
from scipy.io import wavfile
from scipy.interpolate import interp1d
from scipy.signal import butter, filtfilt, iirdesign, zpk2tf, freqz

# the ipython magic below must be commented out in the .py file, since it doesn't work.
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
```

```python
def genTime(freq, length):
    result = []
    timePeriod = 1.0/freq
    time = 0
    for i in xrange(0, length):
        result += [time]
        time += timePeriod
    return result

freq, data = scipy.io.wavfile.read('/Users/vijaytramakrishnan/til/panda.wav')
leftSpeaker = data[:,0]
rightSpeaker = data[:,1]

time = genTime(freq, len(data))

plt.figure()
plt.plot(time,leftSpeaker,'r',label='leftStereo')
plt.xlabel('time (s)')
plt.ylabel('Amplitude')

freqs = np.fft.rfftfreq(len(leftSpeaker), 1.0/freq)
fourier_transform = np.fft.rfft(leftSpeaker)
```

```python
#plot sines
plt.figure()
plt.plot(freqs,fourier_transform.real,'r',label='sine waves')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
```

```python
#plot cosines
plt.figure()
plt.plot(freqs,fourier_transform.imag,'g',label='cosine waves')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
```
